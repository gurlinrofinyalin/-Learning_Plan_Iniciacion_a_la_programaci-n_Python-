#### Creating train and test dataset
Train/Test Split involves splitting the dataset into training and testing sets that are mutually exclusive. After which, you train with the training set and test with the testing set. 
This will provide a more accurate evaluation on out-of-sample accuracy because the testing dataset is not part of the dataset that have been used to train the model. Therefore, it gives us a better understanding of how well our model generalizes on new data.

This means that we know the outcome of each data point in the testing dataset, making it great to test with! Since this data has not been used to train the model, the model has no knowledge of the outcome of these data points. So, in essence, it is truly an out-of-sample testing.

Let's split our dataset into train and test sets. 80% of the entire dataset will be used for training and 20% for testing. We create a mask to select random rows using __np.random.rand()__ function: 


##### Creación de conjuntos de datos de entrenamiento y prueba
La División de Entrenamiento/Prueba (Train/Test Split) implica dividir el conjunto de datos en conjuntos de entrenamiento y prueba que son mutuamente excluyentes. Después de esto, se entrena con el conjunto de entrenamiento y se prueba con el conjunto de prueba.

Esto proporcionará una evaluación más precisa de la exactitud "fuera de muestra" (out-of-sample accuracy), ya que el conjunto de datos de prueba no forma parte del conjunto de datos que se ha utilizado para entrenar el modelo. Por lo tanto, nos da una mejor comprensión de qué tan bien nuestro modelo generaliza con datos nuevos.

Esto significa que conocemos el resultado de cada punto de datos en el conjunto de datos de prueba, ¡lo que lo hace ideal para probar! Dado que estos datos no se han utilizado para entrenar el modelo, el modelo no tiene conocimiento del resultado de estos puntos de datos. Así que, en esencia, es verdaderamente una prueba fuera de muestra.

Dividamos nuestro conjunto de datos en conjuntos de entrenamiento y prueba. El 80% de todo el conjunto de datos se utilizará para el entrenamiento y el 20% para la prueba. Creamos una máscara para seleccionar filas aleatorias usando la función np.random.rand():


````python
# Genera una máscara booleana aleatoria.
# 'np.random.rand(len(df))' crea un array de números aleatorios distribuidos uniformemente entre 0 y 1.
# La longitud de este array es igual al número de filas en el DataFrame original 'df'.
# Al comparar cada número aleatorio con 0.8, se produce un array de valores True o False:
#   - Si el número aleatorio es menor que 0.8, el valor correspondiente en 'msk' será True.
#   - Si el número aleatorio es mayor o igual que 0.8, el valor correspondiente en 'msk' será False.
# Esto resulta en que aproximadamente el 80% de los valores en 'msk' sean True y el 20% sean False,
# lo que se utilizará para dividir los datos.
msk = np.random.rand(len(df)) < 0.8

# Crea el conjunto de datos de entrenamiento ('train').
# Se seleccionan todas las filas del DataFrame 'cdf' donde el valor correspondiente en la máscara 'msk' es True.
# Este conjunto contendrá aproximadamente el 80% de los datos de 'cdf', seleccionados de forma aleatoria.
train = cdf[msk]

# Crea el conjunto de datos de prueba ('test').
# Se seleccionan todas las filas del DataFrame 'cdf' donde el valor correspondiente en la máscara 'msk' es False.
# El operador '~' (tilde) invierte los valores booleanos de la máscara 'msk'.
# Este conjunto contendrá aproximadamente el 20% restante de los datos de 'cdf', también seleccionados aleatoriamente.
# La división en conjuntos de entrenamiento y prueba es fundamental para evaluar cómo el modelo
# se generaliza a datos no vistos.
test = cdf[~msk]
````

### Simple Regression Model
Linear Regression fits a linear model with coefficients B = (B1, ..., Bn) to minimize the 'residual sum of squares' between the actual value y in the dataset, and the predicted value yhat using linear approximation. 


### Modelo de Regresión Simple
La Regresión Lineal ajusta un modelo lineal con coeficientes B = (B1, ..., Bn) para minimizar la "suma residual de cuadrados" entre el valor real 'y' en el conjunto de datos y el valor predicho 'yhat' utilizando aproximación lineal.

#### Train data distribution


````python
# Crea una máscara booleana (True/False) aleatoria.
# np.random.rand(len(df)) genera un array de números aleatorios entre 0 y 1,
# con una longitud igual al número de filas en el DataFrame original 'df'.
# Al comparar cada número con 0.8, se crea la máscara:
# Aproximadamente el 80% de los valores serán True (si el número aleatorio es menor que 0.8)
# y el 20% serán False (si el número aleatorio es igual o mayor que 0.8).
msk = np.random.rand(len(df)) < 0.8

# Crea el conjunto de datos de entrenamiento (train).
# Selecciona todas las filas del DataFrame 'cdf' donde la máscara 'msk' es True.
# Este conjunto contendrá aproximadamente el 80% de los datos originales, seleccionados aleatoriamente.
train = cdf[msk]

# Crea el conjunto de datos de prueba (test).
# Selecciona todas las filas del DataFrame 'cdf' donde la máscara 'msk' es False (usando '~' para negar la máscara).
# Este conjunto contendrá aproximadamente el 20% restante de los datos originales, también seleccionados aleatoriamente.
# Es crucial que los conjuntos de entrenamiento y prueba sean mutuamente excluyentes y aleatorios
# para una evaluación imparcial del modelo.
test = cdf[~msk]
````

#### Modeling
Using sklearn package to model data.


Modelado
Usando el paquete sklearn para modelar datos.


````python
# Importa el módulo 'linear_model' de la librería scikit-learn, que contiene modelos lineales, incluyendo la regresión lineal.
from sklearn import linear_model

# Crea una instancia del modelo de Regresión Lineal.
# 'regr' es ahora un objeto que representa el modelo de regresión lineal simple que vamos a entrenar.
regr = linear_model.LinearRegression()

# Prepara los datos de las características (variable independiente) para el entrenamiento.
# Convierte la columna 'ENGINESIZE' del conjunto 'train' a un array de NumPy.
# Esta será nuestra variable de entrada 'x' para el modelo. Se usa `np.asanyarray`
# para asegurar el formato correcto que espera scikit-learn.
train_x = np.asanyarray(train[['ENGINESIZE']])

# Prepara los datos de la variable objetivo (variable dependiente) para el entrenamiento.
# Convierte la columna 'CO2EMISSIONS' del conjunto 'train' a un array de NumPy.
# Esta será nuestra variable de salida 'y' que el modelo intentará predecir.
train_y = np.asanyarray(train[['CO2EMISSIONS']])

# Entrena el modelo de regresión lineal.
# El método 'fit' calcula los coeficientes (la pendiente de la línea) y el intercepto (el punto donde la línea cruza el eje Y)
# que mejor se ajustan a los datos de entrenamiento (train_x y train_y).
regr.fit(train_x, train_y)

# Imprime los coeficientes calculados por el modelo.
# En una regresión lineal simple, 'regr.coef_' contendrá un solo valor, que es la pendiente de la línea.
print ('Coefficients: ', regr.coef_)

# Imprime el intercepto (o término constante) calculado por el modelo.
# 'regr.intercept_' es el valor de 'y' cuando 'x' es cero.
print ('Intercept: ',regr.intercept_)
````

As mentioned before, __Coefficient__ and __Intercept__ in the simple linear regression, are the parameters of the fit line. 
Given that it is a simple linear regression, with only 2 parameters, and knowing that the parameters are the intercept and slope of the line, sklearn can estimate them directly from our data. 
Notice that all of the data must be available to traverse and calculate the parameters.


Como se mencionó anteriormente, el Coeficiente y la Intercepción en la regresión lineal simple son los parámetros de la línea ajustada.

Dado que es una regresión lineal simple, con solo 2 parámetros, y sabiendo que los parámetros son la intercepción y la pendiente de la línea, scikit-learn puede estimarlos directamente a partir de nuestros datos.

Ten en cuenta que todos los datos deben estar disponibles para recorrerlos y calcular los parámetros.

#### Plot outputs


#### Salidas de los gráficos


We can plot the fit line over the data:


Podemos trazar la línea de ajuste sobre los datos:


````python
# Crea un gráfico de dispersión (scatter plot) utilizando los datos de entrenamiento.
# Los puntos representan la relación entre 'ENGINESIZE' (Tamaño del motor) en el eje X
# y 'CO2EMISSIONS' (Emisiones de CO2) en el eje Y.
# Los puntos del gráfico se mostrarán en color azul.
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')

# Dibuja la línea de regresión (línea de mejor ajuste) sobre el gráfico de dispersión.
# - 'train_x': Son los valores del tamaño del motor usados para entrenar el modelo.
# - 'regr.coef_[0][0]*train_x': Esto calcula la parte de la pendiente (coeficiente) multiplicada por los valores de 'train_x'.
#   'regr.coef_' devuelve un array de arrays, por eso se accede con `[0][0]` para obtener el valor numérico del coeficiente.
# - '+ regr.intercept_[0]': Suma el valor del intercepto (punto de corte con el eje Y) a la ecuación de la línea.
#   'regr.intercept_' también devuelve un array, por eso se accede con `[0]` para obtener el valor numérico.
# - '-r': Es un formato de cadena que indica que la línea debe ser de color rojo ('r') y sólida ('-').
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')

# Establece la etiqueta del eje X (horizontal) del gráfico como "Engine size" (Tamaño del motor).
plt.xlabel("Engine size")

# Establece la etiqueta del eje Y (vertical) del gráfico como "Emission" (Emisión).
plt.ylabel("Emission")

# Nota: El comando plt.show() no está en este fragmento, pero sería necesario para visualizar el gráfico.
````

#### Evaluation
We compare the actual values and predicted values to calculate the accuracy of a regression model. Evaluation metrics provide a key role in the development of a model, as it provides insight to areas that require improvement.

There are different model evaluation metrics, lets use MSE here to calculate the accuracy of our model based on the test set: 
* Mean Absolute Error: It is the mean of the absolute value of the errors. This is the easiest of the metrics to understand since it’s just average error.

* Mean Squared Error (MSE): Mean Squared Error (MSE) is the mean of the squared error. It’s more popular than Mean Absolute Error because the focus is geared more towards large errors. This is due to the squared term exponentially increasing larger errors in comparison to smaller ones.

* Root Mean Squared Error (RMSE). 

* R-squared is not an error, but rather a popular metric to measure the performance of your regression model. It represents how close the data points are to the fitted regression line. The higher the R-squared value, the better the model fits your data. The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).


#### Evaluación
Comparamos los valores reales y los valores predichos para calcular la precisión (exactitud) de un modelo de regresión. Las métricas de evaluación desempeñan un papel clave en el desarrollo de un modelo, ya que proporcionan información sobre las áreas que requieren mejora.

Existen diferentes métricas de evaluación de modelos; usemos el MSE aquí para calcular la precisión de nuestro modelo basándonos en el conjunto de prueba:

Error Absoluto Medio (MAE - Mean Absolute Error): Es la media del valor absoluto de los errores. Esta es la más fácil de entender de las métricas, ya que es simplemente el error promedio.

Error Cuadrático Medio (MSE - Mean Squared Error): El Error Cuadrático Medio (MSE) es la media del error al cuadrado. Es más popular que el Error Absoluto Medio porque el enfoque está más dirigido a los errores grandes. Esto se debe a que el término al cuadrado aumenta exponencialmente los errores más grandes en comparación con los más pequeños.

Raíz del Error Cuadrático Medio (RMSE - Root Mean Squared Error).

R-cuadrado (R-squared) no es un error, sino una métrica popular para medir el rendimiento de tu modelo de regresión. Representa cuán cerca están los puntos de datos de la línea de regresión ajustada. Cuanto mayor sea el valor de R-cuadrado, mejor se ajusta el modelo a tus datos. La mejor puntuación posible es 1.0 y puede ser negativa (porque el modelo puede ser arbitrariamente peor).


````python
# Importa la función 'r2_score' del módulo 'metrics' de scikit-learn.
# Esta función se utiliza para calcular la puntuación R-cuadrado.
from sklearn.metrics import r2_score

# Prepara los datos de las características (variable independiente) del conjunto de prueba.
# Convierte la columna 'ENGINESIZE' del conjunto 'test' a un array de NumPy.
# Estos son los datos de entrada que se pasarán al modelo para hacer predicciones.
test_x = np.asanyarray(test[['ENGINESIZE']])

# Prepara la variable objetivo (variable dependiente real) del conjunto de prueba.
# Convierte la columna 'CO2EMISSIONS' del conjunto 'test' a un array de NumPy.
# Estos son los valores reales con los que compararemos las predicciones del modelo.
test_y = np.asanyarray(test[['CO2EMISSIONS']])

# Realiza predicciones sobre el conjunto de prueba utilizando el modelo 'regr' (ya entrenado).
# 'test_y_' contendrá los valores de emisiones de CO2 predichos por el modelo para 'test_x'.
test_y_ = regr.predict(test_x)

# Calcula y imprime el Error Absoluto Medio (MAE - Mean Absolute Error).
# Mide el promedio de las diferencias absolutas entre los valores predichos (test_y_) y los valores reales (test_y).
# Un MAE más bajo indica que las predicciones del modelo están, en promedio, más cerca de los valores reales.
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))

# Calcula y imprime la Suma de Cuadrados Residuales (RSS - Residual Sum of Squares),
# que en este contexto es equivalente al Error Cuadrático Medio (MSE - Mean Squared Error).
# Mide el promedio de las diferencias al cuadrado entre los valores predichos y los reales.
# Penaliza más fuertemente los errores grandes debido al término al cuadrado. Un MSE más bajo es mejor.
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))

# Calcula y imprime la puntuación R-cuadrado (R2-score).
# Esta métrica indica la proporción de la varianza en la variable dependiente que puede ser predicha por el modelo.
# Un valor más cercano a 1.0 indica un mejor ajuste del modelo a los datos (mayor capacidad de explicación).
# Puede ser negativo si el modelo es peor que simplemente predecir la media.
print("R2-score: %.2f" % r2_score(test_y , test_y_))
````
